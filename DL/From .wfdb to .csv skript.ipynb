```python
from os import listdir, mkdir, system
from os.path import isfile, isdir, join, exists
import os
```

```python
path = '../Research/training2017/'
dir_out = path + 'csv/'

if not exists(dir_out):
    mkdir(dir_out)
    
records = [f for f in listdir(path) if isfile(join(path, f)) if(f.find('.mat') != -1)]
print(records)

os.chdir(path)


for r in records:
    command = 'rdsamp -r ' + r[:-4] + ' -c -H -v >' + r[:-4] + '.csv'
    print(command)
    system(command)
```

```python
import shutil

source = '../Research/training2017/'
destination = '../Research/training2017/csv/'

files = os.listdir(source)

for file in files:
    if file.endswith('.csv'):
        print(file)
        shutil.move(os.path.join(source, file), os.path.join(destination, file))
```

```python
shutil.make_archive('training2017_csv', 'zip', '/home/kate/Research/training2017/csv/')
```

```python
path = '/home/kate/Research/training2017/csv/'
os.chdir(path)

def zipdir(path, ziph):
    # ziph is zipfile handle
    for root, dirs, files in os.walk(path):
        for file in files:
            ziph.write(os.path.join(root, file))

if __name__ == '__main__':
    zipf = zipfile.ZipFile('training2017_csv.zip', 'w', zipfile.ZIP_STORED)
    zipdir('/home/kate/Research/', zipf)
    zipf.close()
```
